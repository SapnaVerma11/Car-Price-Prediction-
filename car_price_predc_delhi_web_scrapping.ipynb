```python
# importing required libraries.
import pandas as pd
import selenium
import time

#importing Webdriver
from selenium import webdriver

#importing Exception Handling
from selenium.common.exceptions import NoSuchElementException
#importing keys
from selenium.webdriver.common.keys import Keys
```

```python
#connecting the web driver.
driver=webdriver.Chrome(r'C:\Program Files (x86)\chromedriver_win32\chromedriver.exe')
```

```python
#connecting the website.
driver.get('https://www.olx.in/delhi_g2001152/cars_c84')
time.sleep(2)
```

```python
#creating empty list
Car_Brand=[]
Car_Model=[]
Car_Variant=[]
Year_of_Manufacturing=[]
Transmission=[]
Fuel_Type=[] 
Kms_Driven=[]
Price=[]
No_of_Owners=[]
```

```python
###Getting the URL of the cars

##creating empty list for product URL

Prod_url=[]

#creating a loop to get the Url for all the cars

for i in range(0,1000):
    url=driver.find_elements_by_xpath("//ul[@class='rl3f9 _3mXOU']/li/a")
    for i in url:
        Prod_url.append(i.get_attribute('href'))
    try:
        next_btn=driver.find_element_by_xpath("//button[@class='rui-39-wj rui-3evoE rui-1JPTg']").click()
    except NoSuchElementException:
        pass
    
for i in Prod_url:
    driver.get(i)
    time.sleep(2)
    
    #getting the car brand
    try:
        Car_Brand.append(driver.find_element_by_xpath("//div[@class='_3_knn']/div/span[2]").text)
    except NoSuchElementException as e:
        Car_Brand.append('-')
        
    #getting the car model
    try:
        Car_Model.append(driver.find_element_by_xpath("//span[@data-aut-id='value_model']").text)
    except NoSuchElementException as e:
        Car_Model.append('-')
    
    #getting the car variant
    try:
        Car_Variant=[].append(driver.find_element_by_xpath("//span[@data-aut-id='value_variant']").text)
    except NoSuchElementException as e:
        Car_Variant=[].append('-')
    
    #getting the year of manufacturing
    try:
        Year_of_Manufacturing.append(driver.find_element_by_xpath("//span[@data-aut-id='value_year']").text)
    except NoSuchElementException as e:
        Year_of_Manufacturing.append('-')
    
    #getting the fuel type
    try:
        Fuel_Type.append(driver.find_element_by_xpath("//span[@data-aut-id='value_petrol']").text)
    except NoSuchElementException as e:
        Fuel_Type.append('-')
    
    #getting the transmission
    try:
        Transmission.append(driver.find_element_by_xpath("//span[@data-aut-id='value_transmission']").text)
    except NoSuchElementException as e:
        Transmission.append('-')
    
    #getting the kms run
    try:
        Kms_Driven.append(driver.find_element_by_xpath("//span[@data-aut-id='value_mileage']").text)
    except NoSuchElementException as e:
        Kms_Driven.append('-')
        
    #getting the number of owners    
    try:
        No_of_Owners.append(driver.find_element_by_xpath("//span[@data-aut-id='value_first_owner']").text)
    except NoSuchElementException as e:
        No_of_Owners.append('-')
    
    #getting the price for the car
    try:
        Price.append(driver.find_element_by_xpath("//section[@class='_2wMiF']/span[1]").text.replace('₹',''))
    except NoSuchElementException as e:
        Price.append('-')
```
